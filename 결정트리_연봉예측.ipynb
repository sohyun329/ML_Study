# 결정트리
- 관측값과 목표값을 연결시켜주는 예측모델로서 나무 모양으로 데이터 분류
- 수많은 트리 기반 모델의 기본 모델
- 각 변수의 특정 지점을 기준으로 데이터 분류
- 장점 :
  - 데이터에 대한 가정이 없어 자유롭게 적용할 수 있다
  - 아웃라이어에 영향을 거의 받지 않는다
  - 트리 그래프를 통해서 직관적으로 이해하고 설명할 수 있다
- 단점 :
  - 트리가 무한정 깊어지면 오버피팅 문제를 야기할 수 있다
  - 다른 트리 기반 모델들에 비해 예측력이 떨어진다
- 유용한 곳 :
  - 종속변수가 연속형 데이터와 범주형 데이터 모두 사용 가능
  - 모델링 결과 시각화할 목적으로 가장 유용
  - 아웃라이어가 문제될 정도로 많을 때 선형 모델보다 좋은 대안이 될 수 있다


## 1. 라이브러리 및 데이터 불러오기, 데이터 확인

## 2. 전처리 : 범주형 데이터

```python
data['class'] = data['class'].map({'<=50K':0, '>50K':1})
```

1. object형 변수 정보 확인

```python
for i in data.columns:
  print(i, data[i].dtype)
```

```python
obj_list = []
for i in data.columns:
  if data[i].dtype == 'object':
    obj_list.append(i)

obj_list
```

2. 전처리할 변수 선별

```python
for i in obj_list:
  print(i, data[i].nunique())
```

4. occupation 변수 처리

```python
data['occupation'].value_counts()
```

5. natvie-country 변수처리

```python
data['native-country'].value_counts()
```

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
numeric_columns.groupby(data['native-country']).mean().sort_values('class')
```

- 트리의 경우, 범주형 데이터를 숫자로 치환하여 모델링해도 괜찮다. ➡️ 연속된 숫자들을 연속적으로 받아들이기보다 일정 구간을 나누어 받아들이기 때문에, 트리가 깊어지면 범주형 변수를 숫자로 바꿔도 상관없다.

```python
country_group = numeric_columns.groupby(data['native-country']).mean()['class']
```

```python
country_group = country_group.reset_index()
```

```python
country_group
```

```python
data = data.merge(country_group, on='native-country',how='left')
```

```python
data
```

```python
data.drop('native-country',axis=1,inplace=True)
data = data.rename(columns={'class_x':'class','class_y':'native-country'})
```

## 3. 전처리 : 결측치 처리 및 더미 변수 변환

```python
data.isna().mean()
```

```python
data['native-country'] = data['native-country'].fillna(-99)
```

- 결측치에 임의의 숫자를 채워주는 것은 선형모델에서 데이터 왜곡을 불러올 수 있으니 주의해야하지만, 트리 기반 모델에서는 큰 문제가 없다.

```python
data['workclass'].value_counts()
```

```python
data['workclass'] = data['workclass'].fillna('Private')
```

- private가 70% 가량 차지하고 있어 결측치 private로 채우기

```python
data['occupation'].value_counts()
```

```python
data['occupation'] = data['occupation'].fillna('Unknown')
```

```python
data = pd.get_dummies(data, drop_first=True)
```

## 4. 모델링 및 평가

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class',axis=1),data['class'],test_size=0.4, random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

## 5. 결정트리
### 5-1. 분류 결정 트리
- 각 노드의 순도가 가장 높은 방향으로 분류
- 순도 : 한 노드 안에 여러 종류가 아닌 한 종류의 목표값만 있는 상태에 대한 지표
- 순도 평가 지표 : 지니 인덱스, 교차 엔트로피
  - 지니 인덱스 : 지니 인덱스가 가장 낮은 값이 나오는 특정 변수의 특정 값을 기준으로 노드를 분류
  - 교차 엔트로피 : 순도가 높을수록 낮은 값 (최대값은 1까지 나올 수 있음)
- 사이킷런의 결정 트리에서는 기본값으로 지니 인덱스 사용  

### 5-2. 회귀 결정 트리
- MSE를 평가 기준으로 이용
- 가장 낮은 MSE값이 나오도록 노드 분류
- 사이킷런의 결정 트리 모델 기본 값 : MSE (MAE, Poisson 등으로 설정 가능)

## 6. 오버피팅 문제
- 훈련셋을 사용하여 지나치게 깊은 트리 모델을 만들면, 시험셋에 대한 예측 떨어지는 경향이 있다.
- 예측 모델이 훈련셋을 지나치게 잘 예측한다면 새로운 데이터를 예측할 때 큰 오차를 유발할 수 있다. ➡️ 오버피팅
  - 오버피팅은 완전히 피할 수 없음. 대신 줄여나가야함
- 시험셋의 실제값과 예측모델에서 상당한 오차가 발생하는 경우 ➡️ 언더피팅
  - 언더피팅 : 과소적합, 모델이 충분히 학습되지 않아 훈련셋에 대해서도 좋은 예측력을 내지 못하는 상황
  - 언더피팅은 무조건 피해야 함
- 편향-분산 트레이드오프
  - 편향 : 독립변수와 종속변수를 모델링한 알고리즘이 적절치 못하거나 중요한 부분을 놓쳐서 제대로 된 예측을 하지 못할 때 높아진다. (언더피팅 예시)
  - 분산 : 훈련셋에 있는 데이터의 노이즈에 의해 발생하는 오차 (오버피팅 예시)
  - 높은 편향 ➡️ 언더피팅 문제, 높은 분산 ➡️ 오버피팅 문제

## 7. 매개변수 튜닝
- 결정트리에서는 트리 깊이가 깊어질수록 (수없이 많은 노드를 분류하여 모델을 만들수록) 오버피팅 발생 가능성이 높다.
- 트리의 깊이를 제한하는 매개변수 제공

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train score: ', accuracy_score(y_train, train_pred),'Test score: ',accuracy_score(y_test, test_pred))
```

```python
model = DecisionTreeClassifier(max_depth=5)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train score: ', accuracy_score(y_train, train_pred),'Test score: ',accuracy_score(y_test, test_pred))
```

```python
model = DecisionTreeClassifier(max_depth=7)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train score: ', accuracy_score(y_train, train_pred),'Test score: ',accuracy_score(y_test, test_pred))
```
